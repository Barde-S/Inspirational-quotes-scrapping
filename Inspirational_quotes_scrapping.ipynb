```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
```

```python
URL = 'http://www.values.com/inspirational-quotes'
r = requests.get(URL)
```

```python
soup = BeautifulSoup(r.content, 'html5lib')
print(soup.prettify())
```

```python
quotes = []

table = soup.find('div', attrs={'id': 'all_quotes'})

for row in table.findAll('div', attrs={'class': 'col-6 col-lg-4 text-center margin-30px-bottom sm-margin-30px-top'}):
    quote = {}
    try:
        quote['theme'] = row.h5.text.strip() if row.h5 else None
        quote['url'] = row.a['href'] if row.a else None
        quote['img'] = row.img['src'] if row.img else None
        if row.img and 'alt' in row.img.attrs:
            alt_text = row.img['alt'].split(" ")
            quote['lines'] = " ".join(alt_text[0:-1]) if len(alt_text) > 0 else None
            quote['author'] = alt_text[-1] if len(alt_text) > 1 else None
        quotes.append(quote)
    except Exception as e:
        print(f"Error processing row: {e}")

filename = 'inspirational_quotes.csv'
with open(filename, 'w', newline='') as f:
    w = csv.DictWriter(f, ['theme', 'url', 'img', 'lines', 'author'])
    w.writeheader()
    for quote in quotes:
        w.writerow(quote)
```
